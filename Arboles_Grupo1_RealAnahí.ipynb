# Árbol Binario

```python
class NodoArbol:
  def __init__(self, clave):
    self.clave = clave
    self.izq = None
    self.der = None

class BST:
  def __init__(self):
    self.raiz = None
  def insertar(self, clave):
    self.raiz= self._insertar(self.raiz, clave)
  def insertar(self, nodo, clave):
    if nodo is None:
      return NodoArbol(clave)
      if clave < nodo.clave:
        nodo.izq = self._insertar(nodo.izq, clave)
      else:
        nodo.der = self._insertar(nodo.der, clave)
      return nodo
  def buscar(self, clave):
    return self._buscar(self.raiz, clave)
  def _buscar(self, nodo, clave):
    if nodo is None or nodo.clave == clave:
      return nodo is not None
    if clave < nodo.clave:
      return self._buscar(nodo.izq, clave)
    else:
      return self._buscar(nodo.der, clave)
```
