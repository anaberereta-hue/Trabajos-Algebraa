##Ejercicio 4: Recursividad Lineal

```python
def factorial(n):
  #Caso base
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)
print(factorial(5))
```

##Ejercicio 5: Recursividad Binaria

```python
def fibonacci(n):
  if n == 0:
    return 0
  elif n == 1:
    return 1
  else:
    return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(5))
```
