##Ejercicio 1: Recursividad Clásica

```python
# ── Factorial ──────────────────────────────
def factorial(n):
    """
    Calcula el factorial de un número usando recursividad.
    La función se llama a sí misma reduciendo el valor de 'n'
    hasta llegar al caso base.
    """
    # Caso base
    if n <= 1: return 1

    # Caso recursivo
    return n * factorial(n - 1)

print(factorial(5)) # 5*4*3*2*1 = 120
```

##Ejercicio 2: Recursividad Clásica

```python
# ── Fibonacci ──────────────────────────────
def fibonacci(n):
    """
    Calcula el n-ésimo número de la secuencia de Fibonacci.
    En esta secuencia, cada número se obtiene sumando los dos anteriores.
    """
    # Caso base
    if n <= 1: return n

    # Caso recursivo
    return fibonacci(n-1) + fibonacci(n-2)
```

##Ejercicio 3: Recursividad Clásica

```python
# ── Suma de arreglos con recursión ─────────────
def suma_array(arr, n):
    """
    Resuelve la suma de una manera muy directa:
    La suma total es igual al último número de la lista
    más la suma de todos los números anteriores.
    """
    # Caso base: Si no hay números para sumar (n es 0)
    if n == 0: return 0

    # Caso recursivo: Tomamos el número actual (arr[n-1]) y le sumamos
    # el resultado de volver a ejecutar la función con los números que quedan.
    return arr[n-1] + suma_array(arr, n-1)

print(suma_array([1, 2, 3, 4, 5], 5)) # Imprime 15
```

##Ejercicio 4: Recursividad Lineal

```python
def factorial(n):
  # Caso base: Si n es 0, el factorial es 1.
  if n == 0:
    return 1
  else:
    # Caso recursivo: n multiplicado por el factorial de (n-1).
    return n * factorial(n-1)
print(factorial(5))
```

##Ejercicio 5: Recursividad Binaria

```python
def fibonacci(n):
  # Caso base: El primer número de Fibonacci (posición 0) es 0.
  if n == 0:
    return 0
  # Caso base: El segundo número de Fibonacci (posición 1) es 1.
  elif n == 1:
    return 1
  else:
    # Caso recursivo: La suma de los dos números anteriores en la secuencia.
    return fibonacci(n-1) + fibonacci(n-2)
print(fibonacci(5))
```

##Ejercicio 6: Recursión de Cola

```python
def factorial_cola(n, acumulador=1):
  # Caso Base: Si n es 0, el acumulador ya tiene el resultado final.
  if n == 0:
    return acumulador

  # Caso recursivo: Se realiza la llamada recursiva como la última operación,
  # actualizando el acumulador con el producto de n.
  return factorial_cola(n - 1, acumulador * n)
print(factorial_cola(5)) # 120
```

##Ejercicio 7: Recursión Múltiple

```python
def suma_triple(n):
  # Caso Base: Si n es menor o igual a 0, se devuelve 1 como condición de parada.
  if n <= 0:
    return 1

  # Caso recursivo: Se suman los resultados de tres llamadas recursivas
  # con n-1, n-2 y n-3.
  return suma_triple(n-1) + suma_triple(n-2) + suma_triple(n-3)
print(suma_triple(4)) # 17
```

##Ejercicio 8: Factorial Recursivo

```python
def factorial_cola(n, acumulador=1):
  # Caso Base: Cuando n llega a 0, se devuelve el acumulador que contiene el resultado final.
  if n ==0:
    return acumulador

  # Caso recursivo: Se llama a la función con n-1 y se actualiza el acumulador
  # multiplicándolo por el n actual. Esto convierte la recursión en recursión de cola.
  return factorial_cola(n-1, acumulador * n)
```

##Ejercicio 9: Factorial Iterativo


```python
def factorial_iterativo(n):
  # Inicializa el resultado en 1, ya que el factorial de 0 es 1 y es el punto de partida.
  resultado = 1

  # Itera desde 1 hasta n (inclusive) para multiplicar cada número.
  for i in range (1, n+1):
    resultado *= i

  # Retorna el factorial calculado de forma iterativa.
  return resultado
```
