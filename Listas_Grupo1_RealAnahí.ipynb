##Ejercicio 1: Lista Simple

```python
class Nodo:
  """Representa un nodo individual en la lista enlazada."""
  def __init__(self, dato):
    self.dato = dato  # El valor que almacena el nodo
    self.siguiente = None # Puntero al siguiente nodo en la lista

class ListaSimple:
  """Gestiona la estructura de una lista enlazada simple."""
  def __init__(self):
    self.cabeza = None # El primer nodo de la lista (inicialmente vacía)

  def insertar_inicio(self, v):
    """Inserta un nuevo nodo al principio de la lista."""
    n = Nodo(v)
    n.siguiente = self.cabeza
    self.cabeza = n

  def insertar_final(self, v):
    """Inserta un nuevo nodo al final de la lista."""
    n = Nodo(v)
    if not self.cabeza:
        self.cabeza = n
        return
    a = self.cabeza
    while a.siguiente:
      a = a.siguiente
    a.siguiente = n

  def imprimir(self):
    """Imprime los elementos de la lista enlazada."""
    a, items = self.cabeza, []
    while a:
      items.append(str(a.dato))
      a = a.siguiente
    print(" → ".join(items))
# Uso rápido
ls = ListaSimple()
[ls.insertar_final(x) for x in [10,20,30]]
ls.imprimir()
```

##Ejercicio 2 : Lista Simple Aplicación Ciencia de Datos


```python
#Caso de uso: Pipeline de Transformación de Datos (ETL)
class NodoPipeline:
    def __init__(self, nombre, transformacion):
        self.nombre = nombre
        self.transformacion = transformacion
        # Funcion callable
        self.siguiente = None

def ejecutar_pipeline(cabeza, datos):
    actual, resultado = cabeza, datos
    while actual:
        resultado = actual.transformacion(resultado)
        print(f"✓ {actual.nombre}: {len(resultado)} registros")
        actual = actual.siguiente
    return resultado

def cargar_csv(d): return d      # Simulación de carga
def limpiar_nulls(d): return d   # Simulación de limpieza
datos_raw = [1, 2, 3]            # Simulación de datos

# Uso: Construir Pipeline dinámicamente
paso1 = NodoPipeline("Carga", cargar_csv)
paso1.siguiente = NodoPipeline("Limpieza", limpiar_nulls)
datos_finales = ejecutar_pipeline(paso1, datos_raw)
```

##Ejercicio 3:  Lista Doblemente Enlazada


```python
class NodoDoble:
    """
    Representa un elemento individual de la lista.
    Almacena el valor y los punteros hacia el nodo previo y el siguiente.
    """
    def __init__(self, dato):
        self.dato = dato
        self.anterior = self.siguiente = None

class ListaDoble:
    """
    Gestiona la estructura de la lista doblemente enlazada,
    manteniendo las referencias al principio (cabeza) y al final (cola).
    """
    def __init__(self):
        self.cabeza = self.cola = None

    def insertar_final(self, valor):
        """
        Crea un nuevo nodo y lo anexa al final de la estructura.
        Si la lista está vacía, el nodo inicializa la cabeza y la cola.
        """
        nuevo = NodoDoble(valor)
        if not self.cola:
            self.cabeza = self.cola = nuevo
        else:
            nuevo.anterior = self.cola
            self.cola.siguiente = nuevo
            self.cola = nuevo

    def eliminar_nodo(self, nodo): # O(1)
        """
        Desvincula un nodo específico aislando sus conexiones.
        Reasigna los punteros de los nodos vecinos o ajusta la cabeza/cola si es necesario.
        """
        if nodo.anterior:
            nodo.anterior.siguiente = nodo.siguiente
        else:
            self.cabeza = nodo.siguiente

        if nodo.siguiente:
            nodo.siguiente.anterior = nodo.anterior
        else:
            self.cola = nodo.anterior

    def inverso(self):
        """
        Imprime los elementos recorriendo la lista en sentido contrario:
        desde la cola retrocediendo hasta llegar a la cabeza.
        """
        actual = self.cola
        while actual:
            print(actual.dato)
            actual = actual.anterior
```

##Ejercicio 4: Lista Doble Aplicación LRU Cache en Ciencia de Datos

```python
# Definiciones
class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.prev = self.siguiente = None

class ListaDoble:
    def __init__(self):
        self.cabeza = self.cola = None

    def insertar_inicio(self, dato):
        nuevo = Nodo(dato)
        if not self.cabeza: self.cabeza = self.cola = nuevo
        else:
            nuevo.siguiente = self.cabeza
            self.cabeza.prev = nuevo
            self.cabeza = nuevo
        return nuevo

    def eliminar_nodo(self, nodo):
        if nodo.prev: nodo.prev.siguiente = nodo.siguiente
        else: self.cabeza = nodo.siguiente
        if nodo.siguiente: nodo.siguiente.prev = nodo.prev
        else: self.cola = nodo.prev

# Simulación de modelo
modelo = type('Modelo', (), {'predict': lambda self, x: f"Pred_de_{x}"})()
features_42 = "user_42"

class LRUCache:
    """Cache LRU usando Lista Doble + dict — O(1) get y put"""
    def __init__(self, capacidad):
        self.cap = capacidad
        self.mapa = {} # clave -> nodo
        self.lista = ListaDoble() # MRU en cabeza, LRU en cola

    def get(self, clave): # O(1)
        if clave not in self.mapa:
            return -1
        nodo = self.mapa[clave]
        self.lista.eliminar_nodo(nodo)
        nuevo = self.lista.insertar_inicio(nodo.dato)
        self.mapa[clave] = nuevo
        return self.mapa[clave].dato["valor"]

    def put(self, clave, valor): # O(1)
        if clave in self.mapa:
            self.lista.eliminar_nodo(self.mapa[clave])

        if len(self.mapa) == self.cap: # desalojar LRU
            del self.mapa[self.lista.cola.dato["clave"]]
            self.lista.eliminar_nodo(self.lista.cola)

        nuevo = self.lista.insertar_inicio({"clave": clave, "valor": valor})
        self.mapa[clave] = nuevo

# Uso en Data Science: cachear predicciones costosas
cache = LRUCache(100)
cache.put("user_42_features", modelo.predict(features_42))
print(cache.get("user_42_features")) # Esto imprime el resultado
```
