```python
# 1. Definimos la clase primero
class NodoPipeline:
    def __init__(self, nombre, transformacion):
        self.nombre = nombre
        self.transformacion = transformacion
        self.siguiente = None

# 2. ¡IMPORTANTE! Definimos las funciones de transformación
# Sin esto, el programa no sabe qué hacer en cada paso.
def cargar_csv(datos):
    print("Cargando datos...")
    return datos  # Aquí iría tu lógica real de carga

def limpiar_nulls(datos):
    print("Limpiando valores nulos...")
    return datos  # Aquí iría tu lógica de limpieza

# 3. Definimos la función que ejecuta todo
def ejecutar_pipeline(cabeza, datos):
    actual, resultado = cabeza, datos
    while actual:
        resultado = actual.transformacion(resultado)
        print(f"✓ {actual.nombre}: {len(resultado)} registros")
        actual = actual.siguiente
    return resultado

# 4. Uso: Ahora sí podemos construirlo porque ya definimos todo arriba
datos_raw = [1, 2, 3, None, 5]  # Un ejemplo de datos iniciales

paso1 = NodoPipeline("Carga", cargar_csv)
paso1.siguiente = NodoPipeline("Limpieza", limpiar_nulls)

datos_finales = ejecutar_pipeline(paso1, datos_raw)
```

```python
# 1. Necesitamos definir qué es un Nodo
class Nodo:
    def __init__(self, dato):
        self.dato = dato
        self.prev = None
        self.siguiente = None

# 2. Necesitamos definir la ListaDoble que usará el Cache
class ListaDoble:
    def __init__(self):
        self.cabeza = None
        self.cola = None

    def insertar_inicio(self, dato):
        nuevo = Nodo(dato)
        if not self.cabeza:
            self.cabeza = self.cola = nuevo
        else:
            nuevo.siguiente = self.cabeza
            self.cabeza.prev = nuevo
            self.cabeza = nuevo
        return nuevo

    def eliminar_nodo(self, nodo):
        if nodo.prev:
            nodo.prev.siguiente = nodo.siguiente
        else:
            self.cabeza = nodo.siguiente
        if nodo.siguiente:
            nodo.siguiente.prev = nodo.prev
        else:
            self.cola = nodo.prev

# 3. Tu clase LRUCache (Corregida con la indentación necesaria)
class LRUCache:
    """Cache LRU usando Lista Doble + dict — O(1) get y put"""
    def __init__(self, capacidad):
        self.cap = capacidad
        self.mapa = {} # clave → nodo
        self.lista = ListaDoble() # MRU en cabeza, LRU en cola

    def get(self, clave): # O(1)
        if clave not in self.mapa:
            return -1
        nodo = self.mapa[clave]
        self.lista.eliminar_nodo(nodo)
        nuevo_nodo = self.lista.insertar_inicio(nodo.dato)
        self.mapa[clave] = nuevo_nodo
        return self.mapa[clave].dato["valor"]

    def put(self, clave, valor): # O(1)
        if clave in self.mapa:
            self.lista.eliminar_nodo(self.mapa[clave])

        if len(self.mapa) == self.cap: # desalojar LRU
            del self.mapa[self.lista.cola.dato["clave"]]
            self.lista.eliminar_nodo(self.cola)

        nuevo = self.lista.insertar_inicio({"clave": clave, "valor": valor})
        self.mapa[clave] = nuevo

# --- Simulación de modelo para que no de error ---
class Modelo:
    def predict(self, x): return f"Predicción para {x}"

modelo = Modelo()
features_42 = "datos_de_usuario"

# 4. Uso en Data Science:
cache = LRUCache(100)
cache.put("user_42_features", modelo.predict(features_42))

print(f"Resultado en cache: {cache.get('user_42_features')}")
```
